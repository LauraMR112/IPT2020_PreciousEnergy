# Polynomials
As a first approximation, we can approximate the power curve by 2 polynomical functions: before and after the maximum. The objective of this process is to have a function that can be integrated to calculate the energy (area under that curve).

```python
def poly1(x,a,b):
    return a*x + b
arg1 = [1,1]
def poly2(x,a,b,c):
    return a*x**2 + b*x + c
arg2 = [1,1,1]
def poly3(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d
arg3 = [1,1,1,1]
def poly4(x,a,b,c, d, e):
    return a*x**4 + b*x**3 + c*x**2 + d*x + e
arg4 = [1,1,1,1,1]
```

Now, the function called Energy does the whole process starting with the measured voltages until the total energy extracted by the turbine is obtained.

```python
def Energy(dataList,time,peakPoint,polynomial1,argument1,curveName='', ymax = 0.052):
    """
    Function to calculate the total energy extracted by the turbine from the measured voltages along time.
    
    Parameters:
    -----------
    dataList : voltages in mV
    time : the total time of the measurement in seconds
    peakPoint : position of the maximum power
    polynomial1 : type of polynomial to fit the first part of the data
    argument1 : guess parameters to fit polynomial1 to the first part of data
    curveName : Title of the power curve plot
    ymax : Optional parameter to set the maximum y limit
    
    Return:
    ----------
    EnergyT : The energy calculated from the area under the power curve
    """ 
    R = 4.2                                            #Turbine resistance
    power = [(i/1000)**2/R for i in dataList]
    timeX = np.linspace(0,time+1,len(dataList))        #List of times of each voltage
    
    power1 = power[0:peakPoint]
    power2 = power[(peakPoint-1):-1]
    timeX1 = timeX[0:peakPoint]
    timeX2 = timeX[(peakPoint-1):-1]

    popt1, pcov1 = curve_fit(polynomial1, timeX1, power1, argument1)  #Fit until the peak
    popt2, pcov2 = curve_fit(poly4, timeX2, power2, arg4)             #Fit after the peak

    fit1 = polynomial1(timeX1,*popt1)
    fit2 = poly4(timeX2,*popt2)
    
    #Integration to obtain the area under the curve
    Int1 = quad(polynomial1, timeX1[0], timeX1[-1],args=(tuple(popt1)))   
    Int2 = quad(poly4, timeX2[0], timeX2[-1], args=(popt2[0],popt2[1],popt2[2],popt2[3],popt2[4]))
    
    EnergyT = Int1[0] + Int2[0]
    EnergyT = round(EnergyT,4)           #Energy extracted by the turbine              
    
    #Plot of power data and the polynonials fitted
    fig=plt.figure(figsize=(8,5))
    plt.title(curveName, fontsize = 18)
    plt.xlabel('Time [seg]', linespacing=1, fontsize = 18)
    plt.ylabel('Power [W]', linespacing=1, fontsize = 18)
    plt.plot(timeX,power,marker='o',linestyle='',color='r', linewidth=2, label='E = '+str(EnergyT)+' [J]')
    plt.plot(timeX1,fit1,linestyle='-',color='k', linewidth=2)
    plt.plot(timeX2,fit2,linestyle='-',color='k', linewidth=2)
    plt.xticks(size = 'x-large')
    plt.yticks(size = 'x-large')
    plt.legend(loc=1, prop={'size': 18})
    plt.xlim(-1,41)
    plt.ylim(0, ymax)
    plt.grid(True)
    plt.close()

    return EnergyT, fig, timeX,power,timeX1,fit1,timeX2,fit2
```
